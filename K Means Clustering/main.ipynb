# K means Clustering UNSUPERVISED LEARNING! -ADVANCED-

```python
# Import libraries
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline

# Import dataset
df = pd.read_csv('dataset.csv')

# Assign X variables
X = df[['total_bill','size']]

# Assign algorithm 
model = KMeans(n_clusters=6)

# Fit algorithm to data
model.fit(X)
#_______________________________________________________________________________________________________________________________________
# Run algorithm
model_predict = model.predict(X)
centroids = model.cluster_centers_
print(model.cluster_centers_) 

# Plot centroids and clusters
plt.figure(figsize=(7,5))
plt.scatter(X['total_bill'], X['size'], c=model_predict, s=50, cmap='rainbow')
plt.scatter(centroids[:, 0], centroids[:, 1], c='black', s=200, alpha=1);
#_______________________________________________________________________________________________________________________________________
```

# Process explained step by step:

<h3>
    
1) Firstly we import the important libraries and a Dataset (matplotlib to see the results). 
2) Then we just assign Variables to Axis. (Y : size, X: total_bill)
3) Then we just assign the algorithm and number of <b>CLUSTERS:</b> <code>model = KMeans(<b>n_clusters=6</b>)</code>
4) After, we fit the algorithm to data: <code> model.fit(X)</code>
5) ### run algorithm
6) ### plot results

</h3>
